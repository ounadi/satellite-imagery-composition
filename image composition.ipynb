## Etape2 Affichage d'une image en couleur 

### Première composition  TM3 ,TM2,TM1

```python
imgR1=Image.new('RGB',(800,400))#créer image pour savegarder le resultat 
for i in range(800):
    for y in range(400): #Boucle pour parcourir tout les pixel de 3 images
        r=img3[y,i]
        v=img2[y,i]
        b=img1[y,i]
        imgR1.putpixel((i,y),(r,v,b))
imgR1.save('reslt1.png','PNG')# savegarde l'image resultat
display(imgR1) #Affichage resultat
```

#### Observation : image est colorisée à partir de ses trois image, qui montre certaines zones, la mer ,les escpases verts


### Deuxième composition TM4,TM2 et TM1

```python
imgR2=Image.new('RGB',(800,400))#créer image pour savegarder la resultat 
for i in range(800):
    for y in range(400):
        r=img4[y,i]
        v=img2[y,i]
        b=img1[y,i]
        imgR2.putpixel((i,y),(r,v,b))
imgR2.save('reslt2.png','PNG')# savegarde l'image resultat
display(imgR2)#Affichage resultat
```

### Observation :
#### On remarque les zones prés de la mer sont bleu et ce qui est a l'intérieur est en rouge (TM4 proche infrarouge), ( bleu c'est tout ce qui humidité et source d'eau ). 

#### Oui il possible de faire d’autres compositions de n'importe quelle image en niveau de gris on peut extraire des couleurs rgb ce qui fait une image en couleur  chaque bande est appliqué pour extraire des informations bien précises (comme il a été mentionné dans le tableau des bandes), comme par exemple TM7,TM2 et TM1

```python
imgR2=Image.new('RGB',(800,400))#créer image pour savegarder la resultat 
img7=cv2.imread('TM7.BMP',0)
for i in range(800):
    for y in range(400):
        r=img7[y,i]
        v=img2[y,i]
        b=img1[y,i]
        imgR2.putpixel((i,y),(r,v,b))
imgR2.save('reslt3.png','PNG')# savegarde l'image resultat
display(imgR2)#Affichage resultat
```
